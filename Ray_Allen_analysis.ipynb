<h1>Effectiveness of NBA data Plot Analysis</h1>

NBA player plays can be explicitly shown by the stat data viusalization. Here a few an examples of how data visualization effectively<br>
Let's take a look at the average weight changes of NBA players throughout the years<br><br>
Import required packages using ``import ~ as ~`` shorting the package name for easier usage and read the data using ``pandas`` package<br>
Sort out the excel data using ``pd.read_excel(file_location)`` that are read with ``.sort_values`` method<br>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


print("\n\n                                            The Chart Below is NBA average stats\n\n")
NBA_stat = pd.read_excel('../data/NBA_stats.xlsx')
NBA_stat_sorted = NBA_stat.sort_values(by='Season', ascending = True)
print(NBA_stat)
print("\n\n                                            The Chart Below is Ray Allen's NBA career stats\n\n")
Allen_stat = pd.read_excel('../data/Ray_Allen_average.xlsx')
print(Allen_stat)
print('\nFirstly I will extract important columns from NBA stats\n')
NBA_want = pd.read_excel('../data/NBA_stats.xlsx')
NBA_want_sorted = NBA_want.sort_values(by='Season', ascending = True)
print(NBA_want)

```

<br>Bring out the wanted values by ``variable[index]``<br><br>

```python
NBA_season = NBA_want_sorted['Season']
NBA_weight = NBA_want_sorted['Wt']
```

<br>Draw the line plot<br>
``plt.figure(size)`` method draws the outline of a plot, ``plt.plot(details)``method helps you to put the wanted data in x,y axis and select the options inside the plot. 
``plt.xlabel(label name)``method sets the lable name of each axis, ``plt.title(title name)``, puts the title of the plot<br> ``plt.grid()``method shows the grid, which is the lines inside the plot<br><br>
Last but not least ``plt.show()`` method is the final code to visualize the plot<br><br>

```python
plt.figure(figsize=(40,20))
plt.plot(NBA_season, NBA_weight, color = 'red', linestyle='dashed', marker = 'o', markerfacecolor ='blue', markersize=10, lw=4)
plt.xlabel('Season', size =20)
plt.ylabel('Player Weight', size = 20)
plt.title('NBA Player Weight Change', size=30)
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

Plot shows almost a steady increase of weight throughout the seasons and a slight decrease in recent years.<br>
Players needed more muscle than lung capacity.<br><br><br><br>
Let's look at other important stat change by creating line plots. <b>Shooting</b><br><br>

<br>Bring out the wanted values by ``variable[index]`` method and draw two seperate line plots<br><br>

```python
NBA_FG_percent = NBA_want_sorted['FG%']
NBA_three_percent = NBA_want_sorted['3P%']

plt.figure(figsize = (80,25))
plt.plot(NBA_season, NBA_FG_percent, lw=10)
plt.xlabel('Season', size =50)
plt.ylabel('FieldGoal Percentage', size = 50)
plt.title('NBA Player FieldGoal Percentage', size=65)
plt.grid()
plt.show()

plt.figure(figsize=(80,25))
plt.plot(NBA_season, NBA_three_percent, color = 'purple', linestyle='-.', marker = 'p', markerfacecolor ='white', markersize=20, lw=13)
plt.xlabel('Season', size =50)
plt.ylabel('3point Percentage', size = 50)
plt.title('NBA Player 3point Percentage', size=65)
plt.grid()
plt.show()
```

Due to NBA Rule alteration and stragic changes, fieldgoal percentage showed a clear drop till late 1990s<br> Now reached an increase for the last 3 seasons.<br>
3point percentage showed steady increase throughout the seasons.<br>

<h1> Animation </h1>

<br>Import the ``patches``, ``path``, ``animation`` tools from ``matplotlib`` package and set the data<br>
using ``variable.iloc[]`` method to sort out wanted values<br><br>

```python
import matplotlib.patches as patches
import matplotlib.path as path
import matplotlib.animation as animation

data = Top_SG_free.drop(['Season'],1)

first_season = data.iloc[0]
second_season = data.iloc[1]
third_season = data.iloc[2]
fourth_season = data.iloc[3]
fifth_season = data.iloc[4]
sixth_season = data.iloc[5]
seventh_season = data.iloc[6]
eighth_season = data.iloc[7]
ninth_season = data.iloc[8]
tenth_season = data.iloc[9]
eleventh_season = data.iloc[10]
twelveth_season = data.iloc[11]
```

<br><br>Set the data using ``numpy.array``<br><br>

```python
ani_data = np.array([
first_season,
second_season,
third_season,
fourth_season,
fifth_season,
sixth_season,
seventh_season,
eighth_season,
ninth_season,
tenth_season,
eleventh_season,
twelveth_season
])
print(ani_data)
```

<br>Make animation executable inside Jupyter Notebook using ``%matplotlib notebook`` code<br><br>

```python
%matplotlib notebook
```

<br>Set the outlines of the animation frame<br><br>

```python
n, bins = np.histogram(ani_data, 12)
left = np.array(bins[:-1])
right = np.array(bins[1:])
bottom = np.zeros(len(left))
top = bottom + n
nrects = len(left)
```

<br>set up the vertex and path codes
arrays using ``plt.Path.MOVETO``, ``plt.Path.LINETO`` and
``plt.Path.CLOSEPOLY`` for each rect.

need 1 ``MOVETO`` per rectangle, which sets the initial point.
need 3 ``LINETO``'s, which tell Matplotlib to draw lines from
vertex 1 to vertex 2, v2 to v3, and v3 to v4.
We then need one ``CLOSEPOLY`` which tells Matplotlib to draw a line from
the v4 to our initial vertex (the ``MOVETO`` vertex), in order to close the
polygon.<br><br>


```python
nverts = nrects *(1+3+1)
verts = np.zeros((nverts, 2))
codes = np.ones(nverts, int) * path.Path.LINETO
codes[0::5] = path.Path.MOVETO
codes[4::5] = path.Path.CLOSEPOLY
verts[0::5, 0] = left
verts[0::5, 1] = bottom
verts[1::5, 0] = left
verts[1::5, 1] = top
verts[2::5, 0] = right
verts[2::5, 1] = top
verts[3::5, 0] = right
verts[3::5, 1] = bottom
```

<br>Make an ``animate`` function, which generates and updates the locations of the vertices for the histogram. 
``patch`` will eventually be a ``Patch`` object.<br><br>

```python
patch = None
def animate(i):
    n, bins = np.histogram(ani_data[:,i], 12)
    top = bottom + n
    verts[1::5,1] = top
    verts[2::5, 1] = top
    return [patch, ]
```

<br>Make the animation plot<br>
Execute the animation, ``interval`` per frame 1000 = 1 second<br><br>

```python
fig, ax = plt.subplots()
barpath = path.Path(verts, codes)
patch = patches.PathPatch(
    barpath, facecolor = 'red', edgecolor = 'black', alpha = 0.5)
ax.add_patch(patch)
ax.set_xlim(left[0], right[-1])
ax.set_ylim(bottom.min(), 7)
ax.invert_xaxis
ani = animation.FuncAnimation(fig, animate, repeat = False, blit = False, interval = 1000)
plt.title('Top10 Shooting Guard Freethrow Proportion per Season', size = 15)
plt.ylabel('Rate Proportion', size = 15)
plt.xlabel('Each Season(per slide)', size = 15)
plt.show()
```

<br>Now, check out Ray Allen's average freethrow percentage rate of each season<br>

```python
Allen_freethrow = Allen_stat['FT%']
print(Allen_freethrow)
```

<br> We can obviously see his FT% comes under 0.85 ~ 0.90 rate proportion from the upper animation chart or even over 0.90<br><br>
Let's look at the actual overall average of each players and Ray Allen. Lastly checking the gap of each player's percentage with Allen<br><br>

```python
print('10 Top SG Freethrow Average\n')
print(Top_SG_free.mean())


print('\n\nRay Allen Freethrow Average\n')
print(Allen_freethrow.mean())


print('\n\nNBA Freethrow Average\n')
print(NBA_stat['FT%'].mean())


print('\n\nAllen / NBA overall freethrow Difference\n')
gap = Allen_freethrow.mean() - NBA_stat['FT%'].mean()
print(gap)


print('\n\nAllen / Top 10 SG freethrow Difference\n')
Allen_freethrow.mean() - Top_SG_free.mean()
```

Allen takes the 1st rank freethrow success rate(0.895) among top 10 best shooting guards. Also shows more than 14% higher percentage in comparison with NBA overall average.
